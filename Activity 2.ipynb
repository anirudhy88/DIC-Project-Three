```r
data$Sex.f <- as.numeric(factor(data$Sex, levels = c("male", "female")))
data$Housing.f <- as.numeric(factor(data$Housing, levels = c("own", "free","rent","NA")))
data$Saving.accounts.f <- as.numeric(factor(data$Saving.accounts, levels = c("moderate", "little","quite rich","rich","NA")))
data$Checking.account.f <- as.numeric(factor(data$Checking.account, levels = c("little", "moderate","rich","NA")))
data$Purpose.f <- as.numeric(factor(data$Purpose, levels = c("radio/TV", "repairs","car","business","furniture/equipment","education","domestic appliances","vacation/others")))
head(data)
```

```r
newHead <- data.frame(data$Age, data$Sex.f, data$Job,
                      data$Housing.f, data$Checking.account.f,
                      data$Credit.amount, data$Duration, data$Purpose.f,
                      data$Credit.Risks)
head(newHead)
colnames(newHead)[1] <- "Age"
colnames(newHead)[2] <- "Sex"
colnames(newHead)[3] <- "Job"
colnames(newHead)[4] <- "Housing"
colnames(newHead)[5] <- "Checking.account"
colnames(newHead)[6] <- "Credit.amount"
colnames(newHead)[7] <- "Duration"
colnames(newHead)[8] <- "Purpose"
colnames(newHead)[9] <- "Credit.Risks"
(newHead)
nrow(newHead)
```

```r
newHead <- newHead[complete.cases(newHead),]

```

```r
for (i in 1:8) {
    diff <- max(newHead[,i]) - min(newHead[,i])
    newHead[,i] <- ((newHead[,i] - min(newHead[,i]))/diff)
}
nrow(newHead)
head(newHead)
```

```r
nrow(newHead)
```

```r
library(class)
samplingRate = 0.1
while  (samplingRate <= 0.9) {
    print(samplingRate)
    nPoints <- nrow(newHead)
    numTestSetLabels <- nPoints * (1 - samplingRate)
    training <- sample(1:nPoints, samplingRate * nPoints, replace = FALSE)
    train <- subset(newHead[training, ], select = c(Age, Sex, Job,
                                                Housing, Checking.account,
                                                Credit.amount, Duration,
                                                Purpose))
    testing <- setdiff(1:nPoints, training)
    test <- subset(newHead[testing, ], select = c(Age, Sex, Job,
                                                Housing, Checking.account,
                                                Credit.amount, Duration,
                                                Purpose))
    cl <- newHead$Credit.Risks[training]
    trueLabels <- newHead$Credit.Risks[testing]
    for (k in 1:20) {
        print(k)
        predictedLabels <- knn(train, test, cl, k)
        numOfIncorrectLabels <- sum(predictedLabels != trueLabels)
        misclassificationRate <- numOfIncorrectLabels/numTestSetLabels
        print(misclassificationRate)
    }
    samplingRate = samplingRate + 0.1;
    print("-------------------------")
}
```

In any case of variable sampling rate, the optimal value of k the one which has the least error rate when run for k values of 1 to 20.
